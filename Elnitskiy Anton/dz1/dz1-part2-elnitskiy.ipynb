```python
import numpy as np
import pandas as pd
```

### Задание №1 - создать датафрймы

```python
authors = pd.DataFrame({'author_id':[1, 2, 3], 'author_name':['Тургенев','Чехов','Островский']},
                      columns = {'author_id','author_name'})
authors
```

```python
book = pd.DataFrame({'author_id':[1, 1, 1, 2, 2, 3, 3], 'book_title':['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'], 'price':[450, 300, 350, 500, 450, 370, 290]},columns = {'author_id', 'book_title', 'price'})
book
```

### Задание 2 - Соединить датафрэймы

```python
authors_price = pd.merge(authors, book, on = 'author_id', how = 'inner')
authors_price
```

### Задание 3 - пять самых дорогих книг

```python
top5 = authors_price.nlargest(5,'price')
top5
```

### Задание 4 - группировка по авторам

```python
authors_stat = authors_price.groupby('author_name').agg({'price':['min','max','mean']})
authors_stat.columns = ["min_price","max_price","mean_price"]
authors_stat = authors_stat.reset_index()
authors_stat
```

### Задание 5 - использование pivot_table

```python
authors_price['cover'] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
authors_price
```

```python
book_info = pd.pivot_table(authors_price, values='price', index=['author_name'],
                         columns=['cover'], aggfunc=np.sum)
book_info.columns.name=''
book_info['мягкая'].fillna(0, inplace = True)
book_info['твердая'].fillna(0, inplace = True)
book_info.reset_index()
```

```python
book_info.to_pickle('book_info.pkl')
new_book_info = pd.read_pickle('book_info.pkl')
new_book_info
```
